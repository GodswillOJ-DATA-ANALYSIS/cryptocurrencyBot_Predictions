### Next, We Defined our moving averages and other indicators. Pandas_ta already have these indicators and we just implemented and used them. Pay note to order of arrangment in order not to get stuck.
#### If you notice we we specified the indicators ADX and SAR

#### (ADX) Average Directional Index, Gives the Strength of a specific momentum. And tells the intensity of bearish or bullish mediums. 
#### (SAR), Parabolic Stop and Reverse Oscillator. This tells positions of support and reverse positions.
#### (SMA), The Simple Moving Averaging. This extracts and counts shifts. The sma and corr were implemented manually at the closing positions.
#### (CORR), The Correlator was added to the SMA


```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['ADX'] = ta.adx(df['High'].shift(1), df['Low'].shift(1), df['Close'].shift(1), length = 14)['ADX_14']
df['SAR'] = ta.psar(df['High'].shift(1), df['Low'].shift(1), df['Close'].shift(1), length = 14)['PSARl_0.02_0.2']
df['SMA'] = df['Close'].shift(1).rolling(window=n).mean()
df['Corr'] = df['Close'].shift(1).rolling(window=n).corr(df['SMA'].shift(1))
```

#### Next Lets Drop Na_values again

```python
df = df.dropna()
```

#### Defining our Previous High, Low and Close positions using our defined values from above.

```python
# Create volume high, low, and close with previous minute's OHLC data
df['Prev_High'] = df['High'].shift(1)
df['Prev_Low'] = df['Low'].shift(1)
df['Prev_Close'] = df['Close'].shift(1)
```

### Defining our Opening and closing signal positions

```python
df['OO'] = df['Open'] - df['Open'].shift(1)
df['OC'] = df['Open'] - df['Prev_Close']
```

### Here, we build our strategy. How?
#### We used the differences of our Open position with a shift and an Open without a shift to create a retracement 

```python
df['Ret'] = (df['Open'].shift(-1)-df['Open'])/df['Open']

# Create a column and assign
for i in range(1, n):
    df[i] = df['Ret'].shift(i)
```

```python
df['Ret']
```

### DATA_Cleaning

##### We selected the Corr feat and returns 1 if greater than the current position or -1 if higher.

```python
df.loc[df['Corr']< -1, 'Corr'] = -1
df.loc[df['Corr']> 1, 'Corr'] = 1
df = df.dropna()
```

#### We have to split our data set.
#### Create a variable split that stores 80% of the length of the dataframe

```python
t = .7
split = int(t*len(df))
split
```

#### Define output signals

#### Create a column by name, 'Signal' and initialize with 0
#### Assign a value of -1 to 'Signal' column for the quantile with the lowest returns
#### Assign a value of 1 to 'Signal' column for the quantile with the highest returns

```python
import warnings
warnings.filterwarnings('ignore')

df['Signals'] = 0

df.loc[df['Ret'] > df['Ret'][:split].quantile(q=0.66), 'signal'] = 1

df.loc[df['Ret'] < df['Ret'][:split].quantile(q=0.34), 'signal'] = -1
```

##### Selecting entry datas

```python
from sklearn.preprocessing import StandardScaler
```

```python

## creating grid parameters by creating new dictionary values
param_grid = {'svc__C': [10,100,1000,10000],
             'svc__gamma': [0.01,0.1,1],
             'svc__kernel': ['rbf']
              }
# GSCV = GridSearchCV(SVC(), param_grid, cv=3, verbose = 3)
# Creating Pipeline

steps = [('scaler', StandardScaler()), ('svc', SVC())]

pipeline = Pipeline(steps)

rcv= RandomizedSearchCV(pipeline, param_grid, cv= TimeSeriesSplit(n_split:=2))
```

#### First we define our SVC params c, gamma and kernel and we pass our step params like so above. We create a pipeline and pass in the steps and input the estimator params into our rcv.
##### note that all the dependencies have already been declared in the beginning.

```python
RandomizedSearchCV(pipeline, param_grid, cv= TimeSeriesSplit(n_splits=2))
```

#### Next we fit our X and Y params into our model using the rcv we just created

```python
 rcv.fit(X.iloc[:split], y.iloc[:split])

## defining params
best_C = rcv.best_params_['svc__C']
best_gamma = rcv.best_params_['svc__gamma']
best_kernel = rcv.best_params_['svc__kernel']
```

#### SVM Algorithm

```python
cls = SVC(C=best_C, kernel=best_kernel, gamma=best_gamma)
```

### Training our Data
#### After training our data we have a C value of 10000 and gamma value of 0.01

```python
ss = StandardScaler()
cls.fit(ss.fit_transform(X.iloc[:split]), y.iloc[:split])
```

### Predicting signals

#### using the predict function we predict our transformed data, at a default signal 0 first.

```python
y_predict = cls.predict(ss.transform(X.iloc[split:]))
df['Pred_Signal'] = 0
```

```python
df['Pred_Signal']
```

```python
y_predict
```

##### We added the predicted signals to the Table.

```python
df.iloc[:split, df.columns.get_loc('Pred_Signal')] = pd.Series(
    cls.predict(ss.fit_transform(X.iloc[:split])).tolist()
)

#Save the predicted vals for the test data
df.iloc[split:, df.columns.get_loc('Pred_Signal')] = y_predict
```
